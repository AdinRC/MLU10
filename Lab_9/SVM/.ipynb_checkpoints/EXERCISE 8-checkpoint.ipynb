All feature values are recoded with four significant digits.

### Task 1 Answer

````python
# Importing common libraries 
import pandas as pd
import numpy as np
from sklearn import metrics
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

import warnings 
warnings.simplefilter(action='ignore', category=FutureWarning)
````

````python
data=pd.read_csv("Breast_Cancer_Data.csv")
````

````python
data.head()
````

````python
y = (data['diagnosis'] == 'M').astype(int)
````

````python
# Set ID as index
data.set_index('id', inplace=True)
````

````python
# Examine numerical and categorical data
numerical_data = data.select_dtypes(include=['float64', 'int64'])
categorical_data = data.select_dtypes(include=['object'])
````

````python
# Convert the diagnosis column to binary
data['diagnosis'] = data['diagnosis'].map({'M': 1, 'B': 0})
````

````python
# Create a bar plot of correlations
correlations = data.corr()['diagnosis'].drop('diagnosis')
correlations.abs().sort_values().plot(kind='bar', figsize=(10, 6))
plt.title('Correlation with Diagnosis')
plt.xlabel('Features')
plt.ylabel('Correlation')
plt.show()
````

````python
# Select the two most correlated features
top_2_features = correlations.abs().nlargest(2).index
X = data[top_2_features]
````

````python
# Scale the selected features using MinMaxScaler
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
X_scaled = scaler.fit_transform(X)
X_scaled_df = pd.DataFrame(X_scaled, columns=top_2_features, index=data.index)
````

````python
# Display results
print("Top 2 Correlated Features:")
print(top_2_features)
````

````python
print("\nScaled Features:")
print(X_scaled_df.head())
````

### Task2 Answer

````python
from sklearn.svm import LinearSVC
# Fit a LinearSVC classifier
from sklearn.model_selection import train_test_split
````

````python
# Define target variable (y)
y = data['diagnosis']
````

````python
# Split the data into training and testing sets (optional, but good practice)
X_train, X_test, y_train, y_test = train_test_split(X_scaled_df, y, test_size=0.3, random_state=42)
````

````python
# Train the LinearSVC model
svc = LinearSVC(random_state=42, max_iter=10000)
svc.fit(X_train, y_train)
````

````python
# Pick 300 samples
X_color = X_scaled_df.sample(300, random_state=42)
y_color = y[X_color.index]
````

````python
# Modify y_color for color-coding
y_color = y_color.map({1: 'red', 0: 'yellow'})
````

````python
# Scatter plot X_color
plt.figure(figsize=(10, 6))
plt.scatter(X_color.iloc[:, 0], X_color.iloc[:, 1], c=y_color, s=50, edgecolor='k')
plt.title("Scatter Plot of Two Features with Color-Coded Classes")
plt.xlabel(X_color.columns[0])
plt.ylabel(X_color.columns[1])
plt.show()
````

````python
# Plot the decision boundary
fig, ax = plt.subplots(figsize=(10, 6))
# Scatter plot the points again
ax.scatter(X_color.iloc[:, 0], X_color.iloc[:, 1], color=y_color, s=50, edgecolor='k')
````

````python
# Generate the decision boundary
x_axis, y_axis = np.arange(0, 1, .005), np.arange(0, 1, .005)
xx, yy = np.meshgrid(x_axis, y_axis)
xx_ravel = xx.ravel()
yy_ravel = yy.ravel()
X_grid = pd.DataFrame([xx_ravel, yy_ravel]).T
# Assign column names to X_grid to match the training data
X_grid.columns = X_train.columns  # Use the same column names as the training data
# Predict the classes for the grid points
y_grid_predictions = svc.predict(X_grid)
y_grid_predictions = y_grid_predictions.reshape(xx.shape)
# Plot the decision surface
fig, ax = plt.subplots(figsize=(10, 6))
ax.contourf(xx, yy, y_grid_predictions, cmap=plt.cm.autumn_r, alpha=.3)
# Scatter plot the points
ax.scatter(X_color.iloc[:, 0], X_color.iloc[:, 1], color=y_color, s=50, edgecolor='k')
plt.title("Decision Boundary of LinearSVC")
plt.xlabel(X_color.columns[0])
plt.ylabel(X_color.columns[1])
plt.show()
````

### Task 3 Answer

````python
from sklearn.model_selection import GridSearchCV
from sklearn.svm import SVC

# Define the parameter grid
param_grid = {
    'C': [0.1, 1, 10, 100],
    'gamma': [1, 0.1, 0.01, 0.001],
    'kernel': ['rbf', 'linear', 'poly']
}
````

````python
# Create GridSearchCV
grid_search = GridSearchCV(SVC(), param_grid, cv=5, verbose=2)
````

````python
# Train using X and y
grid_search.fit(X_train, y_train)
````

````python
# Print the best parameters
print("Best Parameters:", grid_search.best_params_)
````

````python
# Define a function for plotting decision boundaries
def plot_decision_boundary(estimator, X, y):
    # Sample 300 points for visualization
    X_color = X.sample(300, random_state=42)
    y_color = y[X_color.index]
    y_color = y_color.map({1: 'red', 0: 'yellow'})
    
    # Scatter plot the sampled points
    fig, ax = plt.subplots(figsize=(10, 6))
    ax.scatter(X_color.iloc[:, 0], X_color.iloc[:, 1], color=y_color, s=50, edgecolor='k')

    # Generate a grid for decision boundary
    x_axis, y_axis = np.arange(0, 1, .005), np.arange(0, 1, .005)
    xx, yy = np.meshgrid(x_axis, y_axis)
    xx_ravel = xx.ravel()
    yy_ravel = yy.ravel()
    X_grid = pd.DataFrame([xx_ravel, yy_ravel]).T
    X_grid.columns = X.columns  # Match column names
    
    # Predict classes for the grid points
    y_grid_predictions = estimator.predict(X_grid)
    y_grid_predictions = y_grid_predictions.reshape(xx.shape)
    
    # Plot the decision surface
    ax.contourf(xx, yy, y_grid_predictions, cmap=plt.cm.autumn_r, alpha=.3)
    plt.title("Decision Boundary")
    plt.xlabel(X.columns[0])
    plt.ylabel(X.columns[1])
    plt.show()
````

````python
# Call the function with the best estimator
best_estimator = grid_search.best_estimator_
plot_decision_boundary(best_estimator, X_scaled_df, y)
````

### Task 4 Answer

````python
from sklearn.svm import SVC
from sklearn.metrics import classification_report
from sklearn.model_selection import train_test_split

# Split the data
X_train, X_test, y_train, y_test = train_test_split(X_scaled_df, y, test_size=0.3, random_state=42)
````

````python
# Define the SVC model with best parameters from GridSearchCV
best_params = grid_search.best_params_  # Get the best parameters from GridSearchCV
svc_model = SVC(C=best_params['C'], gamma=best_params['gamma'], kernel=best_params['kernel'])
````

````python
# Train the model with training data
svc_model.fit(X_train, y_train)
````

````python
# Predict using the testing data
y_pred = svc_model.predict(X_test)
````

````python
# Print the classification report
print("Classification Report:")
print(classification_report(y_test, y_pred))
````
