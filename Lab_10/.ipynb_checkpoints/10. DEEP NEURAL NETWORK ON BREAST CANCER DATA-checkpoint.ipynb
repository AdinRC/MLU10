# DEEP ARTIFICAL NEURAL NETWORK CLASSIFICATION

```python
import pandas as pd
data=pd.read_csv("data\\Breast_Cancer_Data.csv", index_col=0)
```

```python
from sklearn.preprocessing import LabelEncoder

data.diagnosis=LabelEncoder().fit_transform(data.diagnosis)
```

```python
data
```

```python
X = data.drop('diagnosis', axis=1)
y = data.diagnosis
```

```python
# Splitting the dataset into the Training set and Test set
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler=MinMaxScaler()

X_train=scaler.fit_transform(X_train)
X_test=scaler.transform(X_test)
```

```python
# first neural network with keras make predictions
from numpy import loadtxt
from keras.models import Sequential
from keras.layers import Dense

def model_func():
    model = Sequential()
    model.add(Dense(20, input_dim=X.shape[1], activation='relu'))
    model.add(Dense(10, activation='relu'))
    model.add(Dense(5, activation='relu'))
    model.add(Dense(1, activation='sigmoid'))
    # compile the keras model
    model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
    return model
```

```python
model=model_func()
# fit the keras model on the dataset
hist=model.fit(X_train, y_train, validation_data=(X_test,y_test), epochs=150, batch_size=30, verbose=0)
# make class predictions with the model
predictions = model.predict_classes(X_test)
```

```python
from sklearn.metrics import confusion_matrix, classification_report

print(confusion_matrix(y_test, predictions))
print(classification_report(y_test, predictions))
```

```python
# plot history
import matplotlib.pyplot as plt
plt.plot(hist.history['accuracy'], label='train')
plt.plot(hist.history['val_accuracy'], label='val')
plt.legend()
plt.show();
```

### HANDLE OVERFITTING IN NEURAL NETWORK

#### OPTION 1: REDUCING HIDDEN LAYER

```python
def model_small_layer():
    model = Sequential()
    model.add(Dense(20, input_dim=X.shape[1], activation='relu'))
    model.add(Dense(5, activation='relu'))
    model.add(Dense(1, activation='sigmoid'))
    # compile the keras model
    model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
    return model
```

```python
model=model_small_layer()
# fit the keras model on the dataset
hist_small_layer=model.fit(X_train, y_train, validation_data=(X_test,y_test), epochs=150, batch_size=30, verbose=0)
# make class predictions with the model
predictions = model.predict_classes(X_test)

print(confusion_matrix(y_test, predictions))
print(classification_report(y_test, predictions))
```

```python
plt.plot(hist_small_layer.history['accuracy'], label='train')
plt.plot(hist_small_layer.history['val_accuracy'], label='val')
plt.legend()
plt.show();
```

#### OPTION 2: REGULARIZATION

```python
from tensorflow.keras import regularizers

def model_regularized():
    model = Sequential()
    model.add(Dense(20,kernel_regularizer=regularizers.l1(0.01), input_dim=X.shape[1], activation='relu'))
    model.add(Dense(10,kernel_regularizer=regularizers.l1(0.01), activation='relu'))
    model.add(Dense(5,kernel_regularizer=regularizers.l1(0.01), activation='relu'))
    model.add(Dense(1, activation='sigmoid'))
    # compile the keras model
    model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
    return model
```

```python
model=model_regularized()
# fit the keras model on the dataset
hist_regularized=model.fit(X_train, y_train, validation_data=(X_test,y_test), epochs=150, batch_size=30, verbose=0)
# make class predictions with the model
predictions = model.predict_classes(X_test)

print(confusion_matrix(y_test, predictions))
print(classification_report(y_test, predictions))
```

```python
plt.plot(hist_regularized.history['accuracy'], label='train')
plt.plot(hist_regularized.history['val_accuracy'], label='val')
plt.legend()
plt.show();
```

#### OPTION 3: ADD DROPOUT

```python
#Dropout is one of the important concept in the machine learning. 
#It is used to fix the over-fitting issue. Input data may have some of the unwanted data, usually called as Noise. 
#Dropout will try to remove the noise data and thus prevent the model from over-fitting.
from keras.layers import Dropout

def model_dropout():
    model = Sequential()
    
    model.add(Dense(20, input_dim=X.shape[1], activation='relu'))
    model.add(Dropout(0.2))
    model.add(Dense(10, activation='relu'))
    model.add(Dropout(0.2))
    model.add(Dense(5, activation='relu'))
    model.add(Dense(1, activation='sigmoid'))
    # compile the keras model
    model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
    return model
```

```python
model=model_dropout()
# fit the keras model on the dataset
hist_dropout=model.fit(X_train, y_train, validation_data=(X_test,y_test), epochs=150, batch_size=30, verbose=0)
# make class predictions with the model
predictions = model.predict_classes(X_test)

print(confusion_matrix(y_test, predictions))
print(classification_report(y_test, predictions))
```

```python
plt.plot(hist_dropout.history['accuracy'], label='train')
plt.plot(hist_dropout.history['val_accuracy'], label='val')
plt.legend()
plt.show();
```

### WHAT IF YOU DO NOT SCALE YOUR DATA IN NEURAL NETWORK?

```python
X_train_unscaled, X_test_unscaled, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

```python
model=model_func()
# fit the keras model on the dataset
hist_unscaled=model.fit(X_train_unscaled, y_train, validation_data=(X_test_unscaled,y_test), epochs=150, batch_size=30, verbose=0)
# make class predictions with the model
predictions = model.predict_classes(X_test_unscaled)
```

```python
print(confusion_matrix(y_test, predictions))
print(classification_report(y_test, predictions))
```

```python
# plot history
import matplotlib.pyplot as plt
plt.plot(hist_unscaled.history['accuracy'], label='train')
plt.plot(hist_unscaled.history['val_accuracy'], label='val')
plt.legend()
plt.show();
```
